### 1. Input and Output characteristics of `f_oneway` function (SciPy stats library)

The function `scipy.stats.f_oneway()` is used to perform a **one-way ANOVA (Analysis of Variance)** test. It helps determine whether there are statistically significant differences between the means of three or more independent groups.

#### **Input characteristics:**

* The function accepts **two or more independent samples (groups)** as arguments.
* Each group must be a **1-D array-like structure** such as:

  * Python list
  * NumPy array
  * Pandas Series
* Syntax:

  ```python
  f_oneway(group1, group2, group3, ...)
  ```
* Example:

  ```python
  f_oneway([10, 12, 14], [20, 22, 24], [30, 32, 34])
  ```
* Each group represents a **different category or treatment**.
* Assumptions:

  * Samples are independent of each other
  * Data in each group is approximately normally distributed
  * Homogeneity of variance (equal variances across groups)

#### **Output characteristics:**

* Returns:

  ```
  F-statistic, p-value
  ```

* **F-statistic:**

  * Measures ratio of variance between groups to variance within groups.
  * Higher F value suggests greater difference between group means.

* **p-value:**

  * Probability of observing the result under null hypothesis.
  * Null hypothesis: All group means are equal.

#### **Interpretation:**

* If **p-value < 0.05**, reject null hypothesis → at least one group mean is significantly different.
* If **p-value ≥ 0.05**, fail to reject null hypothesis → no significant difference between group means.

### 2. Results of `sm.stats.anova_lm` function (Statsmodels)

The function `statsmodels.stats.anova_lm()` is used to perform ANOVA on a **fitted linear regression model** created using `OLS (Ordinary Least Squares)`.

#### **Input:**

* A fitted OLS model created using `statsmodels.formula.api.ols`
* Example:

  ```python
  model = ols('y ~ C(group)', data=df).fit()
  sm.stats.anova_lm(model)
  ```

#### **Output:**

It returns a **DataFrame-style ANOVA table** with the following columns:

1. **df (degrees of freedom):**

   * Number of independent values associated with each source of variation.

2. **sum_sq (sum of squares):**

   * Measures total variation explained by each factor and residual error.

3. **mean_sq (mean square):**

   * Calculated as:

     ```
     mean_sq = sum_sq / df
     ```

4. **F-value:**

   * Ratio of variance explained by the model to variance within groups.
   * Higher value indicates stronger effect of the factor.

5. **PR(>F) (p-value):**

   * Probability that observed F-statistic is due to chance.

#### **Interpretation:**

* If **p-value < 0.05**, the factor significantly affects the dependent variable.
* If **p-value ≥ 0.05**, the factor does not significantly influence the outcome.

#### **Key insight:**

* It helps break down variation in data into:

  * Explained variation (due to factors)
  * Unexplained variation (error/residual)

### 3. Role of ANOVA in Data Analytics

ANOVA is a **statistical technique used to compare means of multiple groups simultaneously**. It plays a very important role in data analytics by helping analysts make data-driven decisions.

#### **Major roles of ANOVA:**

1. **Comparison of multiple groups:**

   * Instead of doing multiple t-tests, ANOVA compares all groups at once.

2. **Identifying significant factors:**

   * Helps find which variables significantly affect outcomes.

3. **Reducing error in decision making:**

   * Minimizes Type I error (false positives) compared to repeated tests.

4. **Supporting hypothesis testing:**

   * Used to test whether differences in data are statistically meaningful.

5. **Business and research insights:**

   * Helps in optimizing strategies, processes, and treatments.

#### **Applications in analytics:**

* Marketing campaign performance comparison
* Customer behavior analysis
* A/B testing in product design
* Manufacturing quality control
* Healthcare treatment effectiveness

### 4. When to use One-Way ANOVA (with real-world case)

#### **When to use:**

One-Way ANOVA is used when:

* There is **one independent categorical variable (factor)**.
* The factor has **3 or more groups/levels**.
* The dependent variable is **continuous (numerical)**.
* Goal is to compare means across groups.

#### **Real-world example: Education system**

A school wants to analyze whether different teaching methods affect student exam scores.

* Factor (independent variable): Teaching method

  * Method A (traditional)
  * Method B (online learning)
  * Method C (blended learning)

* Dependent variable: Exam scores

#### **Objective:**

To check whether the average scores differ significantly across teaching methods.

#### **Why ANOVA is used:**

* Instead of comparing each pair of methods using multiple t-tests, one-way ANOVA compares all three at once.
* It tells whether at least one method performs differently.

#### **Decision:**

* If significant → teaching method affects performance.
* If not significant → no difference among methods.

### 5. When to use Two-Way ANOVA (with real-world case)

#### **When to use:**

Two-Way ANOVA is used when:

* There are **two independent categorical variables (factors)**.
* Each factor has two or more levels.
* You want to study:

  * Effect of each factor separately
  * Interaction effect between factors
* Dependent variable is continuous.

#### **Real-world example: Agriculture study**

A researcher wants to study crop yield based on:

* Factor 1: Type of fertilizer

  * Fertilizer A
  * Fertilizer B

* Factor 2: Irrigation level

  * Low water
  * High water

* Dependent variable: Crop yield (kg per field)

#### **Objectives:**

Two-way ANOVA helps determine:

1. Does fertilizer type affect crop yield?
2. Does irrigation level affect crop yield?
3. Is there an **interaction effect** between fertilizer and irrigation?

#### **Why interaction matters:**

* Fertilizer A may work best only under high irrigation.
* Fertilizer B may perform better under low irrigation.

#### **Importance:**

* Helps in understanding combined effects of variables.
* More realistic than one-way analysis in complex systems.
