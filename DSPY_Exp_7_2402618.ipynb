**Q.1** What is need for Hypothesis Testing ?

**Ans:-**
Hypothesis Testing is a fundamental statistical technique used to make decisions about a population based on sample data. In real-world scenarios, it is often impractical or impossible to study an entire population due to constraints such as time, cost, and accessibility. Therefore, hypothesis testing provides a scientific and systematic approach to draw conclusions using sample observations.

The primary need for hypothesis testing arises in the following areas:

**1. Decision Making Under Uncertainty**

In many practical situations, decisions must be made with incomplete information. Hypothesis testing allows us to make rational decisions by evaluating evidence from sample data.

**2. Validation of Assumptions**

Researchers and analysts often propose assumptions or claims regarding population parameters. Hypothesis testing helps in verifying whether these assumptions are statistically valid or not.

**3. Scientific Research**

It plays a crucial role in research by testing theories, models, and experimental results. It ensures that conclusions are not based on random chance but on statistical evidence.

**4. Quality Control and Industrial Applications**

Industries use hypothesis testing to maintain quality standards. For example, testing whether the average weight of a product meets the specified requirement.

**5. Comparative Studies**

It is used to compare two or more groups, such as testing the effectiveness of a new drug versus an existing one.

Thus, hypothesis testing is essential for making logical, data-driven decisions in various fields such as science, engineering, business, and healthcare.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Define the steps
steps = [
    "Step 1: Formulate Hypotheses (H₀, H₁)",
    "Step 2: Select Level of Significance (α)",
    "Step 3: Select Appropriate Test Statistic",
    "Step 4: Collect and Analyze Sample Data",
    "Step 5: Calculate Test Statistic",
    "Step 6: Determine Critical Value or p-value",
    "Step 7: Make Decision (p-value ≤ α or p-value > α)",
    "Step 8: Draw Conclusion"
]

# Setup the figure and axes
fig, ax = plt.subplots(figsize=(10, 12))
ax.set_xlim(0, 10)
ax.set_ylim(0, 12)
ax.axis('off') # Hide the axes

# Define box properties
box_width = 4
box_height = 0.8

# Define starting position
start_x = 3
start_y = 11

# Draw boxes and add text for each step
boxes = []
for i, step_text in enumerate(steps):
    y_pos = start_y - i * (box_height + 0.7)
    if i == 6: # Special handling for decision step (making it slightly wider for better text fit)
        current_box_width = box_width + 1
        current_box_x = start_x - 0.5 # Adjust x to keep it centered
    else:
        current_box_width = box_width
        current_box_x = start_x

    rect = patches.Rectangle((current_box_x, y_pos), current_box_width, box_height,
                             facecolor='#D9EDF7', edgecolor='#3A8DBC', linewidth=2, zorder=2)
    ax.add_patch(rect)
    ax.text(current_box_x + current_box_width / 2, y_pos + box_height / 2, step_text,
            ha='center', va='center', fontsize=10, weight='bold', color='#2F5F7C', wrap=True, zorder=3)
    boxes.append((current_box_x, y_pos, current_box_width, box_height))

# Draw arrows to connect the steps
for i in range(len(boxes) - 1):
    x1, y1, w1, h1 = boxes[i]
    x2, y2, w2, h2 = boxes[i+1]

    # Arrow from bottom center of current box to top center of next box
    ax.annotate('', xy=(x2 + w2/2, y2 + h2), xytext=(x1 + w1/2, y1),
                arrowprops=dict(facecolor='black', edgecolor='black', arrowstyle='->', lw=1.5, mutation_scale=15),
                ha='center', va='center', zorder=1)

# Specific arrows for the decision step (Step 7)
x_decision, y_decision, w_decision, h_decision = boxes[6]
x_conclusion, y_conclusion, w_conclusion, h_conclusion = boxes[7]

# Arrow for Reject H₀ (left path from decision)
# For simplicity, I'm just showing a single path into the conclusion for this diagram
# as the previous text cell describes the two conditions leading to a single conclusion step conceptually.

plt.title('Important Steps in Hypothesis Testing', fontsize=16, weight='bold', pad=20)
plt.tight_layout()
plt.show()
```

**Q.3** What is a student t distribution?

**Ans:-**
The Student’s t-distribution is a probability distribution used in statistical analysis when the sample size is small and the population standard deviation is unknown.

It was introduced by William Sealy Gosset under the pseudonym “Student.”

**Characteristics of t-Distribution:**

* It is bell-shaped and symmetrical, similar to the normal distribution.

* It has heavier tails, which means more variability.

* It depends on degrees of freedom (df = n - 1).

* As sample size increases, the t-distribution approaches the normal distribution.

**Applications:**

* Used in t-tests to estimate population mean.

* Suitable when sample size is less than 30.

* Used when population standard deviation is unknown.

Thus, the t-distribution is an important tool for making reliable inferences from small sample data.

**Conclusion:**

At a 95% confidence level, there is no significant difference between the sample mean and the assumed population mean (19 years). Therefore, we accept the null hypothesis and conclude that the average age of TE IT students is approximately 19 years.

```python
import scipy.stats as stats

data = [18, 19, 20, 19, 21, 18, 19, 20]

t_stat, p_value = stats.ttest_1samp(data, 19)

print("t-statistic:", t_stat)
print("p-value:", p_value)

alpha = 0.05
if p_value <= alpha:
    print("Reject H0")
else:
    print("Accept H0")
```
