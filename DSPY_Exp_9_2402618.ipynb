**Q.1** When do we go for Chi square test ?  

**Ans:-**
The Chi-square (χ²) test is a non-parametric statistical test used to analyze categorical data. It is applied when we want to examine whether there is a significant difference between observed frequencies and expected frequencies.

**Situations where Chi-square Test is Used:**
1. **Testing Goodness of Fit**

* To check whether observed data fits a particular theoretical distribution.

* Example: Whether dice outcomes are equally likely.

2. **Testing Independence**

* To determine whether two categorical variables are related or independent.

* Example: Relationship between gender and product preference.

3. **Testing Homogeneity**

* To compare distributions across different groups.

* Example: Comparing voting patterns in different regions.

**Key Conditions:**

* Data must be categorical (nominal/ordinal)

* Observations must be independent

* Expected frequency in each cell should be ≥ 5 (generally)

**Conclusion:**

We use the Chi-square test when dealing with frequency data and when we want to test relationships or differences between categorical variables.

**Q.3** Explain role of Critical value and p-value in Chi square test.

**Ans:-**

**Critical Value**

* The critical value is obtained from the Chi-square distribution table.

* It depends on:

     * Degrees of freedom (df)

     * Significance level (α)

**Role:**

* It defines the rejection region

* If:

    * χ² (calculated) > χ² (critical) → Reject H₀

    * Otherwise → Accept H₀

**p-value**

* The p-value represents the probability of obtaining the observed result assuming H₀ is true.

**Role:**

* Provides a direct method for decision making.

**Decision Rule:**

* If p ≤ α → Reject H₀

* If p > α → Accept H₀

**Q.5** What is degrees of freedom and how to calculate it while performing chi square test.

**Ans:-**

**Degrees of Freedom (df):**

Degrees of freedom represent the number of independent pieces of information used to calculate a statistic. In simpler terms, it's the number of values in the final calculation of a statistic that are free to vary. It's often related to the number of categories or groups minus any constraints imposed by the data.

**How to Calculate Degrees of Freedom for Chi-square Test:**

The calculation depends on the type of Chi-square test being performed:

1.  **For a Chi-square Test of Goodness of Fit:**
    *   `df = k - 1`
    *   Where `k` is the number of categories or cells in your data.
    *   *Example:* If you're testing whether a die is fair by observing outcomes across 6 faces, then `k = 6`, and `df = 6 - 1 = 5`.

2.  **For a Chi-square Test of Independence or Homogeneity:**
    *   `df = (r - 1) * (c - 1)`
    *   Where `r` is the number of rows in the contingency table and `c` is the number of columns in the contingency table.
    *   *Example:* If you're examining the relationship between gender (2 categories: Male, Female) and product preference (3 categories: A, B, C), your contingency table would have 2 rows and 3 columns. So, `df = (2 - 1) * (3 - 1) = 1 * 2 = 2`.
