**Q.2** In IRIS data set which non graphical EDA is performed ?

**Ans:-**
Non-graphical EDA refers to analyzing data without visualization tools, using statistical and mathematical methods.

**Techniques Used:**

1. **Summary Statistics:**

* Mean: Average value of each feature

* Median: Middle value

* Mode: Most frequent value

* Standard Deviation: Measures variability

* Minimum and Maximum values

2. **Frequency Distribution:**

* Count of each species (Setosa, Versicolor, Virginica)

3. **Correlation Analysis:**

* Measures relationship between numerical features

* Helps identify which features are strongly related

4. **Data Inspection:**

* Checking data types

* Verifying absence of missing values

**Q.3** In  Air Quality data set which graph is used to perform multivariate analysis?

**Ans:-**
Multivariate analysis involves analyzing more than two variables simultaneously.

**Graphs Used:**

1. **Scatter Plot Matrix (Pair Plot):**

* Displays pairwise relationships between multiple variables

* Helps identify trends, clusters, and correlations

2. **Heatmap:**

* Represents correlation matrix using colors

* Helps identify strong and weak relationships among variables

**Most Common Choice:**

* Pair Plot (Scatter Matrix) is widely used for multivariate analysis
